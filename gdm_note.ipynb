```python
#Pivot
pivot_df1 = data.pivot_table(index='env', columns='gtyp_id', values='Ys', aggfunc='mean').reset_index()
pivot_df1.columns.name = None  # Eliminar el nombre de las columnas
print("Primer DataFrame Pivoteado (Ambientes como Filas):")
print(pivot_df1)

pivot_df2 = data.pivot_table(index='gtyp_id', columns='env', values='Ys', aggfunc='mean').reset_index()
pivot_df2.columns.name = None  # Eliminar el nombre de las columnas
print("Segundo DataFrame Pivoteado (Genotipos como Filas):")
print(pivot_df2)
```

```python
#Tablas pivoteadas con funciones de agregacion
agg_functions = {
  'mean': 'mean', # Calcula el valor promedio de los datos. El promedio es la suma de todos los valores dividida por la cantidad de valores.
  'min': 'min', # Encuentra el valor mínimo dentro de los datos. El valor mínimo es el menor número en el conjunto de datos.
  'max': 'max', # Encuentra el valor máximo dentro de los datos. El valor máximo es el mayor número en el conjunto de datos.
  'median': 'median', # Calcula la mediana de los datos. La mediana es el valor medio que separa la mitad superior de la mitad inferior del conjunto de datos.
  'sum': 'sum', # Calcula la suma total de los datos. La suma es la adición de todos los valores en el conjunto de datos.
  'std': 'std', # Calcula la desviación estándar de los datos. La desviación estándar mide la cantidad de variación o dispersión de un conjunto de valores.
  'var': 'var' # Calcula la varianza de los datos. La varianza es la medida de la dispersión que indica cuánto se separan los valores de la media.
}
def create_pivot_tables(df, agg_functions):
    pivot_env_genotype = df.pivot_table(
        index='env',
        columns='gtyp_id',
        values='Ys',
        aggfunc=agg_functions
    ).round(1)

    pivot_genotype_env = df.pivot_table(
        index='gtyp_id',
        columns='env',
        values='Ys',
        aggfunc=agg_functions
    ).round(1)

    return pivot_env_genotype, pivot_genotype_env

desired_agg_funcs = ['mean', 'min', 'max', 'median', 'sum', 'std', 'var']
aggfunc_dict = {'Ys': [agg_functions[func] for func in desired_agg_funcs]}
pivot_env_genotype, pivot_genotype_env = create_pivot_tables(data, aggfunc_dict)

print("Pivot Table - Environment vs. Genotype:")
print(pivot_env_genotype)

print("\nPivot Table - Genotype vs. Environment:")
print(pivot_genotype_env)
```

```python
#Clustering
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
scaler = StandardScaler()
pivot_env_genotype_standardized = pd.DataFrame(scaler.fit_transform(pivot_env_genotype),
                                               index=pivot_env_genotype.index,
                                               columns=pivot_env_genotype.columns)
pivot_genotype_env_standardized = pd.DataFrame(scaler.fit_transform(pivot_genotype_env),
                                               index=pivot_genotype_env.index,
                                               columns=pivot_genotype_env.columns)

def plot_dendrograms_matrix(pivot_data, standardized_data, agg_funcs):
    num_funcs = len(agg_funcs)
    fig, axes = plt.subplots(2, num_funcs, figsize=(5 * num_funcs, 20))

    for i, func in enumerate(agg_funcs):
        pivot_data_func = pivot_data.xs(func, axis=1)
        standardized_data_func = standardized_data.xs(func, axis=1)

        linkage_env_genotype = linkage(pivot_data_func, method='ward')
        dendrogram(linkage_env_genotype, labels=pivot_data_func.index, ax=axes[0, i])
        axes[0, i].set_title(f"No estandarizado - {func}")
        axes[0, i].set_xlabel("Ambiente")
        axes[0, i].set_ylabel("Distancia")

        linkage_env_genotype_std = linkage(standardized_data_func, method='ward')
        dendrogram(linkage_env_genotype_std, labels=standardized_data_func.index, ax=axes[1, i])
        axes[1, i].set_title(f"Estandarizado - {func}")
        axes[1, i].set_xlabel("Ambiente")
        axes[1, i].set_ylabel("Distancia")

    plt.tight_layout()
    plt.show()

desired_agg_funcs = ['mean', 'min', 'max', 'median', 'sum', 'std', 'var']

plot_dendrograms_matrix(pivot_env_genotype, pivot_env_genotype_standardized, desired_agg_funcs)

desired_agg_funcs2 = ['mean', 'min', 'max', 'median']

aggfunc_dict2 = {'Ys': [agg_functions[func] for func in desired_agg_funcs2]}

pivot_env_genotype2, pivot_genotype_env2 = create_pivot_tables(data, aggfunc_dict2)

scaler = StandardScaler()
pivot_env_genotype_standardized2 = pd.DataFrame(scaler.fit_transform(pivot_env_genotype2),
                                               index=pivot_env_genotype2.index,
                                               columns=pivot_env_genotype2.columns)
pivot_genotype_env_standardized2 = pd.DataFrame(scaler.fit_transform(pivot_genotype_env2),
                                               index=pivot_genotype_env2.index,
                                               columns=pivot_genotype_env2.columns)
color_threshold_env_gen = 50000
color_threshold_env_gen_std = 45

# Function to plot dendrogram with more y-axis ticks and larger figure size
def plot_dendrogram(linkage_matrix, labels, title, color_threshold, figsize=(12, 9), yticks=None):
    plt.figure(figsize=figsize)
    dendrogram(linkage_matrix, labels=labels, color_threshold=color_threshold)
    plt.title(title)
    plt.xlabel("Ambiente")
    plt.ylabel("Distancia")

    if yticks is not None:
        plt.yticks(yticks)

    plt.show()


# Clustering jerárquico para DataFrame no estandarizado
linkage_env_genotype2 = linkage(pivot_env_genotype2, method='ward')
# Generate more y-axis ticks
yticks_env_gen = range(0, int(max(linkage_env_genotype2[:, 2])) + 5000, 5000)
plot_dendrogram(
    linkage_env_genotype2,
    pivot_env_genotype2.index,
    "Dendrograma para ambientes y genotipos (no estandarizado)",
    color_threshold_env_gen,
    yticks=yticks_env_gen
)

# Clustering jerárquico para DataFrame estandarizado
linkage_env_genotype_std2 = linkage(pivot_env_genotype_standardized2, method='ward')
# Generate more y-axis ticks
yticks_env_gen_std = range(0, int(max(linkage_env_genotype_std2[:, 2])) + 5, 5)
plot_dendrogram(
    linkage_env_genotype_std2,
    pivot_env_genotype_standardized2.index,
    "Dendrograma para ambientes y genotipos (estandarizado)",
    color_threshold_env_gen_std,
    yticks=yticks_env_gen_std
)
clusters_env_gen_non_std = fcluster(linkage_env_genotype2, color_threshold_env_gen, criterion='distance')
clusters_env_gen_non_std_df = pd.DataFrame({'Environment': pivot_env_genotype2.index, 'Cluster': clusters_env_gen_non_std})

# Generate cluster labels based on the threshold for standardized data
clusters_env_gen_std = fcluster(linkage_env_genotype_std2, color_threshold_env_gen_std, criterion='distance')
clusters_env_gen_std_df = pd.DataFrame({'Environment': pivot_env_genotype_standardized2.index, 'Cluster': clusters_env_gen_std})

print("Clusters for environments and genotypes (non-standardized data):")
print(clusters_env_gen_non_std_df)
print("\nClusters for environments and genotypes (standardized data):")
print(clusters_env_gen_std_df)

color_threshold_gen_env = 14000
color_threshold_gen_env_std = 28

# Function to plot dendrogram with more y-axis ticks and larger figure size
def plot_dendrogram(linkage_matrix, labels, title, color_threshold, figsize=(24, 12), yticks=None, x_label_size=12):
    plt.figure(figsize=figsize)
    dendrogram(linkage_matrix, labels=labels, color_threshold=color_threshold)
    plt.title(title)
    plt.xlabel("Genotipos")
    plt.ylabel("Distancia")

    if yticks is not None:
        plt.yticks(yticks)

    plt.xticks(rotation=90, fontsize=x_label_size)
    plt.show()

# Clustering jerárquico para DataFrame no estandarizado
linkage_gen_enviroment2 = linkage(pivot_genotype_env2, method='ward')
yticks_gen_env = range(0, int(max(linkage_gen_enviroment2[:, 2])) + 1000, 1000)
plot_dendrogram(
    linkage_gen_enviroment2,
    pivot_genotype_env2.index,
    "Dendrograma para genotipos y ambientes (no estandarizado)",
    color_threshold_gen_env,
    yticks=yticks_gen_env,
    x_label_size=10
)

# Clustering jerárquico para DataFrame estandarizado
linkage_gen_enviroment_std2 = linkage(pivot_genotype_env_standardized2, method='ward')
# Generate more y-axis ticks
yticks_gen_env_std = range(0, int(max(linkage_gen_enviroment_std2[:, 2])) + 2, 2)
plot_dendrogram(
    linkage_gen_enviroment_std2,
    pivot_genotype_env_standardized2.index,
    "Dendrograma para genotipos y ambientes (estandarizado)",
    color_threshold_gen_env_std,
    yticks=yticks_gen_env_std,
    x_label_size=10
)

color_threshold_gen_env = 14000
color_threshold_gen_env_std = 28

clusters_non_std = fcluster(linkage_gen_enviroment2, color_threshold_gen_env, criterion='distance')
clusters_non_std_df = pd.DataFrame({'Genotype': pivot_genotype_env2.index, 'Cluster': clusters_non_std})

# Generate cluster labels based on the threshold for standardized data
clusters_std = fcluster(linkage_gen_enviroment_std2, color_threshold_gen_env_std, criterion='distance')
clusters_std_df = pd.DataFrame({'Genotype': pivot_genotype_env_standardized2.index, 'Cluster': clusters_std})
print("Clusters for non-standardized data:")
print(clusters_non_std_df)
print("\nClusters for standardized data:")
print(clusters_std_df)
```

```python
def create_pivot_tables(data):
    pivot_genotype_env = data.pivot_table(
        index='gtyp_id',
        columns='env',
        values='Ys',
        aggfunc='mean'
    ).round(1)

    return pivot_genotype_env

pivot = create_pivot_tables(data)

top_gtyps = {}
bottom_gtyps = {}

for env in pivot:
    sorted_ys = pivot[env].sort_values(ascending=False)
    top_25 = int(len(sorted_ys) * 0.25)
    bottom_25 = int(len(sorted_ys) * 0.25)

    top_gtyps[env] = sorted_ys.head(top_25).index.to_list()
    bottom_gtyps[env] = sorted_ys.tail(bottom_25).index.to_list()

from collections import defaultdict

top_genotype_counts = defaultdict(int)
bottom_genotype_counts = defaultdict(int)

for genotypes in top_gtyps.values():
    for genotype in genotypes:
        top_genotype_counts[genotype] += 1

for genotypes in bottom_gtyps.values():
    for genotype in genotypes:
        bottom_genotype_counts[genotype] += 1

top_genotype_counts = dict(top_genotype_counts)
bottom_genotype_counts = dict(bottom_genotype_counts)


sorted_top = dict(sorted(top_genotype_counts.items(), key=lambda item: item[1], reverse=True))
sorted_bottom = dict(sorted(bottom_genotype_counts.items(), key=lambda item: item[1], reverse=True))

final_top = {}
final_bottom = {}

for k,v in sorted_top.items():
    if v >= 10:
        final_top[k] = v

for k,v in sorted_bottom.items():
    if v >= 10:
        final_bottom[k] = v

top_top_gtyps = sorted_top.keys()

clusters_top = clusters_std_df[clusters_std_df['Genotype'].isin(top_top_gtyps)]

print(clusters_top['Cluster'].value_counts())
```
